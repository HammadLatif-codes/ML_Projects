```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python

# Xavier Initialization
def xavier_init(size):
    in_dim = size[0]
    return tf.keras.initializers.GlorotUniform(seed=None)(shape=size)

# Generator Model
def generator_model():
    inputs = tf.keras.Input(shape=[100])
    x = tf.keras.layers.Dense(128, activation='relu')(inputs)
    outputs = tf.keras.layers.Dense(784, activation='sigmoid')(x)
    return tf.keras.Model(inputs=inputs, outputs=outputs)

# Discriminator Model
def discriminator_model():
    inputs = tf.keras.Input(shape=[784])
    x = tf.keras.layers.Dense(128, activation='relu')(inputs)
    outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
    return tf.keras.Model(inputs=inputs, outputs=outputs)

# Function to display generated images
def show_generated_images(generator, epoch, examples=16, dim=(4, 4), figsize=(5, 5)):
    noise = tf.random.normal([examples, 100])
    generated_images = generator(noise)
    generated_images = generated_images.numpy().reshape(examples, 28, 28)

    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(generated_images[i], interpolation='nearest', cmap='gray')
        plt.axis('off')
    plt.suptitle(f"Generated images at epoch {epoch+1}")
    plt.show()

# Create models
generator = generator_model()
discriminator = discriminator_model()

# Separate optimizers for generator and discriminator
discriminator_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
generator_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)

# Loss function
cross_entropy = tf.keras.losses.BinaryCrossentropy()

# Training loop
(X_train, _), (_, _) = mnist.load_data()
X_train = X_train.reshape(-1, 784).astype('float32') / 255.0

batch_size = 128
epochs = 1000
for epoch in range(epochs):
    for i in range(len(X_train) // batch_size):
        # Sample noise and get generated images
        noise = tf.random.normal([batch_size, 100])
        generated_images = generator(noise)

        # Train the discriminator
        real_images = X_train[i*batch_size:(i+1)*batch_size]
        x = tf.concat([real_images, generated_images], axis=0)
        y = tf.concat([tf.ones([batch_size, 1]), tf.zeros([batch_size, 1])], axis=0)

        with tf.GradientTape() as tape:
            d_loss_real = cross_entropy(tf.ones_like(discriminator(real_images)), discriminator(real_images))
            d_loss_fake = cross_entropy(tf.zeros_like(discriminator(generated_images)), discriminator(generated_images))
            d_loss = d_loss_real + d_loss_fake

        gradients_of_discriminator = tape.gradient(d_loss, discriminator.trainable_variables)
        discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

        # Train the generator
        noise = tf.random.normal([batch_size, 100])
        with tf.GradientTape() as tape:
            generated_images = generator(noise)
            g_loss = cross_entropy(tf.ones_like(discriminator(generated_images)), discriminator(generated_images))

        gradients_of_generator = tape.gradient(g_loss, generator.trainable_variables)
        generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
      # Display images generated by the generator every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}, D Loss: {d_loss}, G Loss: {g_loss}")
        show_generated_images(generator, epoch)


    print(f"Epoch {epoch+1}, D Loss: {d_loss}, G Loss: {g_loss}")
```
